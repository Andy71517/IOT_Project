```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# Plotting
plt.figure(figsize=(12, 6))

# Plot 1: PDR vs. Range for different configurations
for sf in df['SF'].unique():
    subset = df[df['SF'] == sf]
    plt.plot(subset['Range'], subset['PDR'], marker='o', label=f'SF {sf}')

plt.title('PDR vs. Range for different Spreading Factors')
plt.xlabel('Range')
plt.ylabel('PDR')
plt.legend()
plt.show()

# Plot 2: PDR vs. Frequency for different configurations
plt.figure(figsize=(12, 6))

for sf in df['SF'].unique():
    subset = df[df['SF'] == sf]
    plt.plot(subset['Freq'], subset['PDR'], marker='o', label=f'SF {sf}')

plt.title('PDR vs. Frequency for different Spreading Factors')
plt.xlabel('Frequency')
plt.ylabel('PDR')
plt.legend()
plt.show()

# Plot 3: PDR distribution for different configurations
plt.figure(figsize=(12, 6))

for sf in df['SF'].unique():
    subset = df[df['SF'] == sf]
    plt.hist(subset['PDR'], bins=10, alpha=0.5, label=f'SF {sf}')

plt.title('PDR Distribution for different Spreading Factors')
plt.xlabel('PDR')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
